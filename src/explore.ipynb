 AnÃ¡lisis de variables multivariante

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(2, 2, figsize = (10, 7))

# Crear un diagrama de dispersiÃ³n mÃºltiple
sns.regplot(ax = axis[0, 0], data = total_data, x = "TOT_POP", y = "Total Hospitals (2019)")
sns.heatmap(total_data[["Total Hospitals (2019)", "TOT_POP"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0], cbar = False)
sns.regplot(ax = axis[0, 1], data = total_data, x = "Population Aged 60+", y = "Total Hospitals (2019)").set(ylabel=None)
sns.heatmap(total_data[["Total Hospitals (2019)", "Population Aged 60+"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 1])

# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()
```
